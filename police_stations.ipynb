# Load the Data

```python
# Import two libraries that are typically useful.

import numpy
import pandas
```

```python
# Read the data in and assign a name to it.
# Note that this is the case where the data is in *the same directory the Notebook is in*.
csv_data = pandas.read_csv('./Police_Stations.csv')

# Show the data
csv_data
```

# Do the Data Analysis
Everything in this section *is not part of actually reading the data in*, but is an example of something you can do once the data is imported

```python
# Get Latitude and Longitude out
lat = csv_data['LATITUDE']
lon = csv_data['LONGITUDE']
```

```python
# These lines allow plots to appear in the notebook, and
# improve how the notebook deals with libraries (by automatically reloading the library if something changes).
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
# Import Libraries for plotting
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Make a plot of where all the police stations are located (latitude and longitude).

# Make the figure itself
fig = plt.figure( figsize=(12,11) )
ax = plt.gca()

# Plot the stations
ax.scatter(
    lon,
    lat,
    s = 80,
)

# Label the axes
ax.set_xlabel('Longitude', fontsize=22, )
ax.set_ylabel('Latitude', fontsize=22, )
ax.set_aspect( 'equal' )
```
